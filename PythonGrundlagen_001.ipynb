## Strukturierung 
Bitte die Quelle berücksichtigen/durcharbeiten
Quelle: https://www.python-kurs.eu/python3_bloecke.php

Python wählt einen pragmatischen Ansatz - statt einrücke nur für die "optic" zu berücksichtigen übernimmt es die Funktion von Klammerstrukturen - das spart Zeit und zwingt einen zugleich "schön lesbaren" Code zu schreiben.


Das folgende Programm muss nicht mathematisch verstanden werden, es soll nur zeigen wie man einrücken kann um "Strucktur" und bedeutung zu geben.  Es berechnet die Pyxtagoreischen Tripel also die Zahlen wie im folgennden Dreieck für die gilt 3² + 4² = 5² (9+16=25)

Quelle und für die die es wissen wollen: https://de.wikipedia.org/wiki/Pythagoreisches_Tripel 

```python
# Funtkion für die Pythagorean triples
from math import sqrt

n = input("Maximal Number? ")
print ("Deine eingabe war: " + n)
n = int(n)+1
for a in range(1,n):
    for b in range(a,n):
        c_square = a**2 + b**2
        c = int(sqrt(c_square))
        if ((c_square - c**2) == 0):
            print(a, b, c)
```

UM eine echte "tiefe" Kopie zu haben muss man eine Funktion namens deep Copy nutzen



```python
# Echte Kopie und diese wird angepasst
from copy import deepcopy
Person1 = ["Swen", ["Seestrasse", "Konstanz"]]

# Mal mit dem "=" und Person2[0]="Sarah" testen, dann gibts nur noch Sarah
#Person2 = Person1
#Person2[0]="Sarah"
print(Person1,Person2)

# Änderungen werden nun bei beiden Personen sichtbar
Person2 = deepcopy(Person1)
print(Person1,Person2)

# Änderungen werden nun nur bei Person2 sichtbar
Person2[0] = "Sarah" # hier mal "Sarah2" inschreiben und genau beobachten was passiert, min 2 mal ausführen!
Person2[1][0] = "Bücklestrasse"
print(Person1,Person2)
```

## Bedingte Anweisungen

Bedingte Anweisungen sind eine einfache und übersichtliche Möglichkeit, Programme zu entwickeln, die entweder nach einer bestimmten Bedingung oder einer anderen führen. Diese Bedingungen können verschiedene "Anweisungen" enthalten, beispielsweise ein Wert aus einer Variablen Ausgeben oder eine verwerfen.
Quelle: https://www.python-kurs.eu/python3_bedingte_anweisungen.php


```python
person = input("Nationalität? ")
if person == "french" or person == "French":
    print("Préférez-vous parler français?")
elif person == "italian" or person == "Italian":
    print("Preferisci parlare italiano?")
else:
    print("Du bist weder Italiener noch Franzose.")
    print("Deshalb reden wir jetzt deutsch!")
```

Etwas abweichend, wenn man eher Typische Eingaben prüft und darauf reagiert ist "Case" 
https://www.python-kurs.eu/struktureller_musterabgleich.php
Hier etwas sperrig "typischer Musterabgleich" genannt.
Man kann hier vordefinierte Werte setzen und darauf Prüfen:

```python
#def definiert eine Methode - praktisch, so kann man das Programmierte wieder benutzen (schon mal fast der Anfang für Objektorientierung)
def greeting(language):
    chosen_language = language.capitalize()
    match chosen_language:
        case 'English':
            print('Hello')
        case 'German':
            print('Hallo!')
        case _:
            print(f"Hallo, ich kenne {chosen_language} nicht")
            
greeting('english')
greeting('french')
```

### While Schleife
While Schleifen führen etwas aus, solange eine Bedingung "wahr" ist.
Quelle: https://www.python-kurs.eu/python3_schleifen.php



```python
n = 100

sum_of_numbers = 0
i = 1
while i <= n:
    sum_of_numbers += i
    i += 1

result = "Summe von 1 bis " + str(n) + ": " + str(sum_of_numbers)
print(result)
```

### for Schleife
Wird i.d.R. genutzt wenn man über z.B. eine liste laufen oder "iterieren" möchte.
hier hilft dann auch oft die range Funktion, schauen wir uns das mal an:
Quelle https://www.python-kurs.eu/python3_for-schleife.php


```python
edibles = ["ham", "spam", "eggs","nuts"]
spam = False

for food in edibles:
    if food == "spam":
        print("No more spam please!")
        spam = True
        continue
    print("Great, delicious " + food)
    # here can be the code for enjoying our food :-)
else: # das Else gehört zum FOR nicht dem if!!
    if spam:
        print("I didn't like the spam!")
    else:
        print("I am so glad: No spam!")
print("Finally, I finished stuffing myself")
```

und Range?? Hilft uns wenn wir ad hoc eine "Liste" erstellen möchten, siehe hier:

```python
n = 100

s = 0
for counter in range(1, n+1):
    s += counter

print("Sum of 1 until " + str(n) + ": " + str(s) )
```

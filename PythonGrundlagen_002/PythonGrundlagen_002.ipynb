Wir kenne bestimmt noch Dinge aus der Mathematik/Physik, wo man mit mehr als einem Wert etwas bestimmen konnte - wenn nicht die "Exponenten" w√§ren ein Beispiel. Meist hat man eine Basis und einen Exponenten (https://www.mathebibel.de/potenzgesetze).
Wir k√∂nnten ja schreiben 4¬≤ oder 6¬≥ oder eben allgemein x<sup>n</sup> und dann eben x die Basis und n der Exponent.
Wenn wir die Programmieren, haben wir also zwei Werte, die wir nutzen/wissen m√ºssten f√ºr eine Funktion "potenz".

```python
# Unsere Potenzfunktion - hinweis - f√ºr so was gibt es Profi L√∂sungen in den Mathematik und Co Bibliotheken, i.d.R. immer schauen,
# ob jemand uns die Arbeit nicht schon abgenommen hat - das Spart Zeit und meist auch Fehler. Hier gehts aber ums lernen des Prinzips
# deswegen eben mal von Hand ;-)
# Unsere Funktion definieren
def potenz(basis, exponent):
    p = 1
    while (exponent > 0):
        p = p * basis
        exponent = exponent -1 # es fehlt der Teil f√ºr negateive Exponenten - nur so am Rande

    return p                    # das Return liefert den Wert von p zur√ºck, da wo die Funktion aufgerufen wird steht am ende also dieser Wert

# Aufruf der Funktion und die Parameter gleich als input nehmen und mit int in einen Integer als Zahl wandeln, sonst gehts nicht
# da er Text nicht zum Rechnen nutzen kann und Ausgeben
print( potenz( int(input("Basis bitte: ")), int(input("Exponent bitte: ")) ) )


# Das sieht Schritt f√ºr Schritt so aus:
ba = input("Basis bitte: ")             # Abfragen
ex = input("Exponent bitte: ")

ba_i = int(ba)                          # Text sicher zu einer ganzzahl machen (Integer)
ex_i = int(ex)

resultat = potenz(ba_i,ex_i)            # Funktion aufrufen, diese liefert uns den Wert

print(resultat)                         # Ausgabe der Variablen mit dem Wert
```

Hinweis: wir haben hier ja extra eine Funktion verrauscht und versuchen dies nun so gut wie es geht wieder linear hinzubekommen.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. NumPy: Generierung von Beispieldaten
np.random.seed(0)
X = np.linspace(0, 10, 100)  # 100 Datenpunkte von 0 bis 10
y = 3 * X + 1 + np.random.normal(0, 1, 100)  # Lineare Funktion mit Rauschen

# 2. SciPy: Kurvenanpassung
# Definiere die Funktion, die wir anpassen m√∂chten (y = a * X + b)
def linear_func(X, a, b):
    return a * X + b

# F√ºhre die Kurvenanpassung durch
params, _ = curve_fit(linear_func, X, y)
a_scipy, b_scipy = params
print(f"SciPy Kurvenanpassung: a = {a_scipy}, b = {b_scipy}")

# 3. scikit-learn: Lineare Regression
# Um die Daten in das richtige Format zu bringen, reshape X f√ºr scikit-learn
X_reshaped = X.reshape(-1, 1)  # scikit-learn erwartet 2D Arrays

# Trainiere ein lineares Regressionsmodell
model = LinearRegression()
model.fit(X_reshaped, y)

# Vorhersage und Modellbewertung
y_pred = model.predict(X_reshaped)
mse = mean_squared_error(y, y_pred)
print(f"scikit-learn Modell: a = {model.coef_[0]}, b = {model.intercept_}")
print(f"Mean Squared Error: {mse}")

# 4. Plot der Ergebnisse
plt.scatter(X, y, label='Daten', color='blue')
plt.plot(X, linear_func(X, a_scipy, b_scipy), label='SciPy Fit', color='red')
plt.plot(X, y_pred, label='scikit-learn Modell', color='green')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Vergleich: SciPy Kurvenanpassung vs. scikit-learn Modell')
plt.show()
```

Also drei Dinge: Klassen, Objekte(Instanzen) und Methoden:
* Klassen sind Baupl√§ne, nach denen werden Objekte, auch als Instanzen bezeichnet, erstellt Beispiel Bauplan VW Golf - Bestellung von Herrn Mayer einen VW Golf GTI in Blau mit CD Payer  ;-)

* Methoden sind Funktionen die man in der Klasse hinterlegt, die alle Objekte auch haben


```python
# Unsere Klasse Mensch, Sie definiert die Methode und Variablen die wir nutzen k√∂nnen  
class Mensch:
    def SageHallo(self):                # Eine Methode die Hallo Sagt und auf eine Variable des Objektes zugreifen soll
        print("Hallo, mein Name ist " + self.name)
    def SetzeNamen(self, n):            # Eine Methode, in ihr setzen sir eine Objektspezifische Variable auf einen Wert
        self.name = n
    def SetzeBaujahr(self, baujahr):    # noch eine Methode, in ihr setzen wir das baujahr f√ºr das Objekt
        self.baujahr = baujahr
  
# Wir nutzen die Klasse und erstellen daraus einen Menschen - also ein Objekt der Klasse     
x = Mensch()
x.SetzeNamen("Marc")    # Mit den Methoden (Funktionen in einer Klasse) k√∂nnen wir Werte setzen
x.SetzeBaujahr(1979)

y = Mensch()            # Noch ein Mensch, aber eben ein ganz anderer oder eben ein anderes Objekt
y.SetzeNamen("Candido")
y.SetzeBaujahr(1993)

x.SageHallo()           # Wir fragen das Objekt x nach seinen Daten
y.SageHallo()           # Wir fragen das Objekt y nach seinen Daten

print(x)            # noch sieht man hier die Object ID
print(y)            # und hier ebenso aber eine Andere !!! 
```

Klassen habe ein paar besondere Methoden diese sind in Python mittels __ am Anfang und Ende gekennzeichnet.
Man kann mit Ihne z.B. festlegen was beim erstellen eines Objekts aus der Klasse (oben ist das genau "x=Mensch()" ) passieren soll, also wie die Klasse initialisiert werden muss.
Dies geht mit einer ganz besondere Methode <code>def __init_(self, name, baujahr):</code><br>
damit muss man nun beim erzeugen Werte f√ºr name und baujahr angeben, es Funktioniert wie eine Methoden oder Funktionsaufruf.

```python
class Mensch:
    def __init__(self, name, baujahr:int): # Methode zur Initialisierung - wird immer automatisch aufgerufen beim erstellen
        self.name = name
        self.baujahr = baujahr
    def SageHallo(self):                
        print("Hallo, mein Name ist " + self.name)
    def SetzeNamen(self, n):            
        self.name = n
    def SetzeBaujahr(self, baujahr):   
        self.baujahr = baujahr

z = Mensch("Maier", 1999) # Ruft automatisch die init auf, ich kann nun nicht mehr ohne die Argumente f√ºr die Parameter weiter machen!
z.SageHallo()
print(z)
```
